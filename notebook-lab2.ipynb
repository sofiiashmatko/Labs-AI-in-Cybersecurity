# Крок 2. Using Pandas to Get Familiar With Your Data

```python
import pandas as pd

melbourne_file_path = '/kaggle/input/datasets/sofiiashmatko/melb-data/melb_data.csv'
melbourne_data = pd.read_csv(melbourne_file_path)

print("Перші 5 рядків датасету:")
print(melbourne_data.head())

print("\nОпис датасету:")
print(melbourne_data.describe())
```

# Крок 3. Selecting Data for Modeling

```python
print("Назви колонок:")
print(melbourne_data.columns)

melbourne_data = melbourne_data.dropna(axis=0)
```

# Вибір цільової зміної та вибір ознак

```python
y = melbourne_data.Price

melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = melbourne_data[melbourne_features]

print("Опис ознак:")
print(X.describe())

print("\nПерші 5 рядків:")
print(X.head())
```

# Building Your Model

```python
from sklearn.tree import DecisionTreeRegressor

# створення моделі
melbourne_model = DecisionTreeRegressor(random_state=1)

# навчання моделі
melbourne_model.fit(X, y)

print(melbourne_model)
```

# Making Predictions

```python
print("Дані для перших 5 будинків:")
print(X.head())

print("\nПрогнозовані ціни:")
print(melbourne_model.predict(X.head()))
```

# Крок 4. Model Validation

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

melbourne_model = DecisionTreeRegressor(random_state=1)
melbourne_model.fit(train_X, train_y)

val_predictions = melbourne_model.predict(val_X)

from sklearn.metrics import mean_absolute_error

print("Середня абсолютна помилка:")
print(mean_absolute_error(val_y, val_predictions))
```

# Крок 5. Underfitting and Overfitting

```python
from sklearn.metrics import mean_absolute_error

def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return mae

for max_leaf_nodes in [5, 50, 500, 5000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print(f"Max leaf nodes: {max_leaf_nodes} \t Mean Absolute Error: {my_mae}")
```
